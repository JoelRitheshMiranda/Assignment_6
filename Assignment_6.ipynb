```python
# This assignment is for visualization using matplotlib:

# data to use:
# url= https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv
# titanic = pd.read_csv(url)
# Charts to plot:
# 1. Create a pie chart presenting the male/female proportion
# 2. Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

titanic = pd.read_csv("https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv")
titanic.head()
```

```python
# 1. Create a pie chart presenting the male/female proportion

piedata = titanic['sex'].value_counts()
plt.pie(piedata, labels=['Male','Female'], shadow=False, startangle=90, autopct='%1.1f%%', colors=['#4285F4', '#DB4437'])
plt.show()
```

```python
# 2. Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

titanic_clean = titanic.dropna(subset=['age'])
x_male = titanic_clean[titanic_clean['sex'] == 'male']["age"]
y_male = titanic_clean[titanic_clean.sex == 'male']["fare"]

x_female = titanic_clean[titanic_clean['sex'] == 'female']["age"]
y_female = titanic_clean[titanic_clean.sex == 'female']["fare"]

fig = plt.figure(figsize=(15,10))
plt.scatter(x_male, y_male, label='Male', alpha=0.7, color='green', marker='o')
plt.scatter(x_female, y_female, label='Female', alpha=0.7, color='red', marker='s')
plt.title('Fare vs Age')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.legend()
plt.show()
```

```python
# It happens all the time: someone gives you data containing malformed strings, Python, lists and missing data. 
# How do you tidy it up so you can get on with the analysis?
# Take this monstrosity as the DataFrame to use in the following puzzles:
# df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN', 'londON_StockhOlm', 'Budapest_PaRis', 'Brussels_londOn'],
# 'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
# 'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
# 'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )', '12. Air France', '"Swiss Air"']})

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.DataFrame({
    'From_To': ['LoNDon_paris', 'MAdrid_miLAN', 'londON_StockhOlm', 'Budapest_PaRis', 'Brussels_londOn'],
    'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
    'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
    'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )', '12. Air France', '"Swiss Air"']
})
df
```

```python
# 1. Some values in the the FlightNumber column are missing. These numbers are meant to
#    increase by 10 with each row so 10055 and 10075 need to be put in place. Fill in these missing
#    numbers and make the column an integer column (instead of a float column).

df['FlightNumber'].fillna(0, inplace=True)
df['FlightNumber'] = df['FlightNumber'].astype(int)
for i in range(1, len(df["FlightNumber"])):
    df.loc[i,'FlightNumber'] = df.loc[i-1,'FlightNumber'] + 10
df
```

```python
# 2. The From_To column would be better as two separate columns! Split each string on the
#    underscore delimiter _ to give a new temporary DataFrame with the correct values. Assign the
#    correct column names to this temporary DataFrame.

temp_df = pd.DataFrame()
temp_df['From'] = df['From_To'].apply(lambda x: x.split("_")[0])
temp_df['To'] = df['From_To'].apply(lambda x: x.split("_")[1])
temp_df
```

```python
# 3. Notice how the capitalisation of the city names is all mixed up in this temporary DataFrame.
#    Standardise the strings so that only the first letter is uppercase 
#    (e.g. "londON" should become "London".)

temp_df['From'] = temp_df['From'].apply(lambda x: x.title())
temp_df['To'] = temp_df['To'].apply(lambda x: x.title())
temp_df
```

```python
# 4. Delete the From_To column from df and attach the temporary DataFrame from the previous questions.

df.drop('From_To', axis=1, inplace=True)
df = pd.concat([temp_df, df], axis=1)
df
```

```python
# 5. In the RecentDelays column, the values have been entered into the DataFrame as a list. 
#    We would like each first value in its own column, each second value in its own column, and so on. 
#    If there isn't an Nth value, the value should be NaN.
#    Expand the Series of lists into a DataFrame named delays, rename the columns delay_1,
#    delay_2, etc. and replace the unwanted RecentDelays column in df with delays.

delays = pd.DataFrame(df["RecentDelays"].values.tolist(), columns=["delay_1", "delay_2", "delay_3"])
delays
```

```python
df.drop('RecentDelays', axis=1, inplace=True)
df = pd.concat([delays, df], axis=1)
df = df[["From","To","FlightNumber","delay_1","delay_2","delay_3","Airline"]]
df
```
